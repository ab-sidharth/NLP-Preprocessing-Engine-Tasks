## Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love"
In NLP, text is treated as tokens.  
"Love" and "love" are considered different because of case sensitivity.  
This increases the vocabulary size and may confuse the model.  
So, we usually convert all text to lowercase during preprocessing.

### 2. What happens if stopwords are not removed?
If stopwords are not removed:
- The dataset will contain many less meaningful words  
- Vocabulary size becomes large  
- Important words may get less importance  
- Model performance may decrease due to noise  

### 3. Two cases where removing stopwords is harmful

**Example 1: Sentiment Analysis**  
"I am not happy" -- If we remove "not", the meaning changes completely.

**Example 2: Question Answering**  
"What is the time?" -- Removing "what" removes the question intent.

### 4. Stemming vs Lemmatization

**Stemming**
- Removes suffixes without considering meaning  
- May produce incomplete words  
- Example: "playing" -- "play"

**Lemmatization**
- Uses dictionary and linguistic rules  
- Produces meaningful root words  
- Example: "better" -- "good"

Lemmatization is more accurate but slower than stemming.

```python
# Import required libraries
import re
import string
from collections import Counter
```

```python
# Task 2: Advanced Preprocessing Function
def preprocess_text(text):
    """
    This function cleans raw text and returns:
    1. cleaned tokens
    2. cleaned sentence
    """

    # Error handling
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = text.lower() # to lowercase
    text = re.sub(r'http\S+|www\S+', '', text) # remove URLs
    text = re.sub(r'\S+@\S+', '', text) # remove email
    text = re.sub(r'\d+', '', text) # remove numbers
    text = re.sub(r'(.)\1{2,}', r'\1', text) # reduce repeated characters
    text = text.translate(str.maketrans('', '', string.punctuation)) # remove punctuation
    text = re.sub(r'[^\w\s]', '', text)  # remove emojis / symbols
    text = re.sub(r'\s+', ' ', text).strip() # Remove extra spaces
    tokens = text.split() # tokenization
    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]
    clean_sentence = " ".join(filtered_tokens)

    return filtered_tokens, clean_sentence
```

```python
# Task 3: Stress Testing
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    print("Original Text :", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)
```

```python
# Task 4: Token Analytics
analysis_data = []
for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = (
        sum(len(token) for token in tokens) / total_tokens
        if total_tokens > 0 else 0
    )
    print("Sentence:", sentence)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_token_length, 2))
    print("-" * 50)
```

### Task 4 Analysis

- After preprocessing, only one meaningful token ("got") remained.

**Sentence that retained the most meaningful tokens:**  
- "I absolutely looooved this product 😍😍" retained the most meaningful tokens after cleaning.

```python
# Task 5: Frequency Analysis
from collections import Counter
all_tokens = []
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

word_freq = Counter(all_tokens)
print("Top 10 Most Frequent Words:")
print(word_freq.most_common(10))
print("\nTop 5 Least Frequent Words:")
print(word_freq.most_common()[-5:])
```

```python
# Task 6: Full Pipeline
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []
    for text in text_list:
        tokens, cleaned = preprocess_text(text)

        all_tokens.append(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Testing full pipeline
pipeline_output = full_pipeline(test_sentences)
print(pipeline_output)
```
